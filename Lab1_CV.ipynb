> # Writing the image(meaning save it to the device after creating it)

* Creating random array 
* then creating\reshaping gray, colored image


```python
random_byte_arr = bytearray(os.urandom(120000))
flat_np_arr = np.array(random_byte_arr)

gray_img = flat_np_arr.reshape(300,400)
bgr_img = flat_np_arr.reshape(100,400,3)
cv.imwrite('grayImgae.png',gray_img)
cv.imwrite('BGR or colored image.png',bgr_img)
```

* first line code, we created an array of 120000 random values ranges from 0 to 255 (byte value)
* we then converted to an array by numpy

* third line is reshaping to gray sclae 300*400 and next line to color 24bit colored image
* save it to the device
