```python
import numpy as np
import cv2 as cv
```

```python
temp_img = cv.imread('circle.jpg')
img = cv.resize(temp_img,(0,0),fx=0.25, fy=0.25)
cv_imshow(img)
```

```python
img_line = cv.line(img, (0,0), (255,255), (155,155,0), 1)
cv_imshow(img_line)
```
