```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt #not used
```

```python
img = cv.imread('doll.jpg')
img = cv.resize(img,(0,0),fx=0.25, fy=0.25)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv_imshow(img)
```

##Feature Dedection using Fast
* we should know that keypoints = Interest points = features
-- interest points or keypoints feature 
*         


```python
#creating an object from fast class with it's default values
fast = cv.FastFeatureDetector_create()
#extract and detect all the keypoints/features in the image
extracted_Keypoints = fast.detect(img, None)
#finially, draw this keypoints/features an the original image
img2 = cv.drawKeypoints(img, extracted_Keypoints, None, color= (255,0,0))
#extracting all default values that comes with fast object and with nonmaxsuppression is ture
print("threshold: {}".format(fast.getThreshold()) )
print("nonMaxSuppression: {}".format(fast.getNonmaxSuppression()) )
print("neighbours: {}".format(fast.getType()) )
print("keypoints with nonmaxSuppressin: {}".format(len(extracted_Keypoints)) )
cv_imshow(img2)

#disable for non-maxSuppression 
fast.setNonmaxSuppression(0)
extracted_Keypoints = fast.detect(img, None)
img3 = cv.drawKeypoints(img, extracted_Keypoints, None, color=(255,0,0))
cv_imshow(img3)
```
